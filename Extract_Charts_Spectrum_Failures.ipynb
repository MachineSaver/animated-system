```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image = Image.open("C:\\Users\\machine_saver\\Downloads\\Vibration_Charts.png")

# Convert the image to a numpy array
image_np = np.array(image)

# Calculate the size of each individual graph
height, width, _ = image_np.shape
graph_height = height // 8
graph_width = width // 5

# Handle the special case graph that spans two rows
merged_graph_height = graph_height * 2

# Extract the individual graphs
graphs = []
for i in range(8):
    for j in range(5):
        if i == 6 and j == 2:  # Special case for the graph that spans two rows
            graphs.append(image_np[i*graph_height:(i+2)*graph_height, j*graph_width:(j+1)*graph_width])
        elif i == 7 and j == 2:  # Skip the lower part of the merged graph
            continue
        else:
            graphs.append(image_np[i*graph_height:(i+1)*graph_height, j*graph_width:(j+1)*graph_width])

# Display the extracted graphs
fig, axs = plt.subplots(8, 5, figsize=(15, 24))
axs = axs.ravel()

for i, graph in enumerate(graphs):
    axs[i].imshow(graph)
    axs[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
def set_plotly_theme(name='Machine Saver Inc.'):
  pio.templates["draft"] = go.layout.Template(
    layout_annotations=[
        dict(
            name="draft watermark",
            text=name,
            textangle=-30,
            opacity=0.1,
            font=dict(color="white", size=50),
            xref="paper",
            yref="paper",
            x=0.5,
            y=0.5,
            showarrow=False,
          )
      ]
  )
  pio.templates.default = "plotly_dark+draft"

set_plotly_theme()
```
